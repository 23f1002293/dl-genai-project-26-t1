```python
# Question 1
# You have a batch of grayscale spectrogram images represented as a PyTorch tensor with shape (32, 1, 64, 64). You need to flatten this tensor (excluding the batch dimension) to feed it into a Fully Connected (Linear) layer. What is the size of the input features for that Linear layer?

import torch

tensor = torch.randn(32, 1, 64, 64)

flattened = torch.flatten(tensor, start_dim=1)

input_features = flattened.shape[1]
print(f"Number of input features: {input_features}")
```

```python
# Question 2
# You are converting a 1-second audio clip (Sample Rate = 16,000 Hz) into a Mel-Spectrogram using torchaudio. The configuration is n_fft=400, hop_length=160, n_mels=64. Based on your code output, what is the resulting tensor shape?

import torch

L = 16000
n_fft = 400
hop_length = 160

waveform = torch.randn(1, L)


stft_out = torch.stft(waveform, n_fft=n_fft, hop_length=hop_length, center=True, return_complex=True)


n_mels = 64
time_steps = stft_out.shape[2]

print(f"Final Mel-Spectrogram shape: (1, {n_mels}, {time_steps})")
```

```python
# Question 4 
# You have a hidden layer defined as nn.Linear(in_features=128, out_features=64). How many total learnable parameters (weights + biases) does this specific layer contain?

import torch
import torch.nn as nn

layer = nn.Linear(in_features=128, out_features=64)

num_weights = layer.weight.numel()
num_biases = layer.bias.numel()
total_params = num_weights + num_biases

print(f"Total learnable parameters: {total_params}")
```

```python
# Question 5
# Your training dataset consists of 1,050 audio files. You initialize a DataLoader with batch_size=32 and drop_last=False. How many batches will run in one epoch?

import math

dataset_size = 1050
batch_size = 32
drop_last = False

num_batches = math.ceil(dataset_size / batch_size)
print(f"Number of batches: {num_batches}")
```

```python
# Question 5 
# You are performing a 3-class classification. The model outputs logits [2.5, 1.0, 0.1] for a target of Class 0. Using nn.CrossEntropyLoss, what is the calculated loss value (rounded to 3 decimals)?

import torch
import torch.nn as nn

# Model outputs (logits)
logits = torch.tensor([[2.5, 1.0, 0.1]])

# Target class
target = torch.tensor([0])

# Define the loss function
criterion = nn.CrossEntropyLoss()

# Calculate the loss
loss = criterion(logits, target)

print(f"Loss value: {loss.item():.4f}")
```

```python
# Question 6
# You are training with SGD. Current Weight = 0.5, Gradient = 0.2, Learning Rate = 0.01. What is the new weight value after one optimizer step?

weight = 0.5
gradient = 0.2
lr = 0.01

new_weight = weight - (lr * gradient)
print(f"New Weight: {new_weight}")
```

```python
# Question 9 
# After training, you evaluate your model on a test set of 200 spectrograms and it correctly predicts 170 of them. What is the accuracy?

total_samples = 200
correct_predictions = 170

accuracy = (correct_predictions / total_samples) 
print(f"Accuracy: {accuracy}%")
```
