```python

import os
import torch
import json
import numpy as np
import random
import pandas as pd
import wandb
from torch.utils.data import Dataset
from transformers import ASTFeatureExtractor, ASTForAudioClassification, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from dotenv import load_dotenv
from tqdm import tqdm
import torchaudio
```

```python

MODEL_ID = "MIT/ast-finetuned-audioset-10-10-0.4593"

PREPROCESSOR_CONFIG = {
    "do_normalize": True,
    "feature_extractor_type": "ASTFeatureExtractor",
    "feature_size": 1,
    "max_length": 1024,
    "mean": -4.2677393,
    "num_mel_bins": 128,
    "padding_side": "right",
    "padding_value": 0.0,
    "return_attention_mask": False,
    "sampling_rate": 16000,
    "std": 4.5689974
}

IS_KAGGLE = os.path.exists("/kaggle/input")
if IS_KAGGLE:
    DATA_DIR = "/kaggle/input/jan-2026-dl-gen-ai-project/messy_mashup"
    WORKING_DIR = "/kaggle/working"
    OUTPUT_DIR = "/tmp/model_checkpoints"
else:
    DATA_DIR = "/Volumes/Disk1/Janani/DLProject/messy_mashup"
    WORKING_DIR = "/Volumes/Disk1/Janani/DLProject/results"
    OUTPUT_DIR = os.path.join(WORKING_DIR, "checkpoint_results")

SUBMISSION_PATH = os.path.join(WORKING_DIR, "submission.csv")

EPOCHS = 15 
BATCH_SIZE = 16 
LEARNING_RATE = 3e-5 # Slightly lower for deeper convergence
SUBSET_SIZE = 0
NUM_WORKERS = 4 # Use all 4 CPU cores to mix audio while GPU trains
TARGET_DURATION = 10 

USE_WANDB = True
WANDB_PROJECT = "messy-mashup-ast"

if IS_KAGGLE:
    import socket
    def is_online():
        try:
            socket.create_connection(("8.8.8.8", 53), timeout=2)
            return True
        except: return False
    
    if is_online():
        os.environ["WANDB_MODE"] = "online"
        print("Internet detected. WandB will sync live.")
    else:
        os.environ["WANDB_MODE"] = "offline"
        print("No internet detected. WandB will log locally to /kaggle/working/wandb")
    
    os.environ["WANDB_DIR"] = "/tmp"
```

```python
class MashupSimulatorDataset(Dataset):
    def __init__(self, stems_list, stems_data, genre_indices, noise_cache, num_variants=1):
        self.stems_data = stems_data # List of tensors [4000, 1024, 128]
        self.genre_indices = genre_indices # {g_idx: [int_idx1, int_idx2]}
        self.noise_cache = noise_cache
        self.num_variants = num_variants
        
        self.samples_map = []
        for g_idx, indices in self.genre_indices.items():
            label_tensor = torch.tensor(g_idx, dtype=torch.long)
            for idx in indices:
                for _ in range(num_variants):
                    self.samples_map.append((idx, label_tensor))
        
        random.shuffle(self.samples_map)

    def __len__(self):
        return len(self.samples_map)

    def __getitem__(self, idx):
        stem_idx, label_tensor = self.samples_map[idx]
        
        w_base = random.uniform(0.7, 1.0)
        mixed = self.stems_data[stem_idx] * w_base
        
        pool = self.genre_indices[label_tensor.item()]
        num_distractors = random.randint(1, 3) 
        if len(pool) > 1:
            other_indices = random.sample(pool, min(num_distractors, len(pool)))
            for o_idx in other_indices:
                if o_idx != stem_idx:
                    # Randomly weight the overlapping tracks
                    w_other = random.uniform(0.3, 0.6)
                    mixed = mixed + (self.stems_data[o_idx] * w_other)
        
        if self.noise_cache:
            noise = self.noise_cache[random.randint(0, len(self.noise_cache)-1)]
            w_noise = random.uniform(0.05, 0.2)
            mixed = mixed + (noise * w_noise)

        if random.random() < 0.6:
            # Frequency mask
            f_size = random.randint(10, 25)
            f_start = random.randint(0, 128 - f_size)
            mixed[:, f_start:f_start+f_size] = -10.0
            
        if random.random() < 0.6:
            # Time mask
            t_size = random.randint(40, 80)
            t_start = random.randint(0, 1024 - t_size)
            mixed[t_start:t_start+t_size, :] = -10.0

        return {
            "input_values": mixed,
            "labels": label_tensor
        }
```

```python
def get_files(data_dir):
    genre_data = {}
    
    search_path = os.path.join(data_dir, "genres_stems")
    if not os.path.exists(search_path):
        search_path = data_dir
        
    print(f"Searching for genre folders in: {search_path}")
    
    genres = sorted([d for d in os.listdir(search_path) 
                   if os.path.isdir(os.path.join(search_path, d)) 
                   and not d.startswith('.') 
                   and d not in ['ESC-50-master', 'checkpoint_results', 'results']])

    for g in genres:
        genre_path = os.path.join(search_path, g)
        stems = []
        for root, _, files in os.walk(genre_path):
            for file in files:
                if file.lower().endswith('.wav'):
                    stems.append(os.path.join(root, file))
        if stems:
            genre_data[g] = stems

    noise_files = []
    esc_dir = os.path.join(data_dir, "ESC-50-master", "audio")
    if os.path.exists(esc_dir):
        noise_files = [os.path.join(esc_dir, f) for f in os.listdir(esc_dir) if f.lower().endswith('.wav')]

    print(f"Found {len(genre_data)} genres and {len(noise_files)} noise files.")
    return genre_data, noise_files
```

```python
def preload_audio(file_list, target_sr, max_duration=10):
    cache = {}
    target_samples = int(target_sr * max_duration)
    print(f"Preloading {len(file_list)} files...")
    
    for path in tqdm(file_list):
        try:
            try:
                waveform, sr = torchaudio.load(path)
            except:
                import librosa
                audio_np, sr = librosa.load(path, sr=None, mono=True)
                waveform = torch.from_numpy(audio_np).unsqueeze(0)

            if sr != target_sr:
                resampler = torchaudio.transforms.Resample(sr, target_sr)
                waveform = resampler(waveform)
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            
            audio = waveform.squeeze(0).numpy()
            
            # Standardize length
            if len(audio) > target_samples:
                audio = audio[:target_samples]
            elif len(audio) < target_samples:
                audio = np.pad(audio, (0, target_samples - len(audio)))
            
            cache[path] = audio.astype(np.float16)
        except Exception as e:
            continue
    return cache
```

```python

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average="macro")
    return {"accuracy": acc, "f1": f1}
```

```python
def init_tokens():
    try:
        from kaggle_secrets import UserSecretsClient
        user_secrets = UserSecretsClient()
        
        try:
            hf_token = user_secrets.get_secret("HF_TOKEN")
            os.environ["HF_TOKEN"] = hf_token
            print("HF token loaded.")
        except: 
            print("HF_TOKEN secret not found.")
            hf_token = None

        if USE_WANDB:
            try:
                wandb_key = user_secrets.get_secret("WANDB_API")
                os.environ["WANDB_API_KEY"] = wandb_key
                try:
                    wandb.login(key=wandb_key, relogin=True)
                    print("WandB authenticated.")
                except:
                    print("WandB offline mode enabled (No Internet).")
            except Exception as e:
                print(f"WandB API key not found. Logging locally only.")
        return hf_token
    except ImportError:
        load_dotenv()
        return os.getenv("HF_TOKEN")
```

```python
def setup_model(genres, id2label, label2id, device, hf_token=None):
    feature_extractor = ASTFeatureExtractor.from_pretrained(
        MODEL_ID, **PREPROCESSOR_CONFIG, token=hf_token
    )
    model = ASTForAudioClassification.from_pretrained(
        MODEL_ID,
        num_labels=len(genres),
        id2label=id2label,
        label2id=label2id,
        ignore_mismatched_sizes=True,
        token=hf_token
    ).to(device)
    return feature_extractor, model
```

```python

def prepare_simulation_data(genre_data, noise_files, feature_extractor):
    cache_file = os.path.join(WORKING_DIR, "stem_features_v3_fast.pt")
    
    if os.path.exists(cache_file):
        print(f"Loading fast cache from {cache_file}...")
        try:
            return torch.load(cache_file)
        except: pass

    print("Step 1: Computing features for all stems...")
    all_genre_files = [p for paths in genre_data.values() for p in paths]
    raw_audio_cache = preload_audio(all_genre_files, feature_extractor.sampling_rate, TARGET_DURATION)
    
    stem_paths = sorted(raw_audio_cache.keys())
    stem_tensors = []
    
    for path in tqdm(stem_paths, desc="Extracting features"):
        audio = raw_audio_cache[path]
        inputs = feature_extractor(audio.astype(np.float32), sampling_rate=feature_extractor.sampling_rate, return_tensors="pt")
        # Store as float32 to eliminate training-time casting
        stem_tensors.append(inputs.input_values.squeeze(0).to(torch.float32))

    noise_tensors = []
    if noise_files:
        raw_noise = preload_audio(noise_files, feature_extractor.sampling_rate, TARGET_DURATION)
        for audio in tqdm(raw_noise.values(), desc="Extracting noise"):
            inputs = feature_extractor(audio.astype(np.float32), sampling_rate=feature_extractor.sampling_rate, return_tensors="pt")
            noise_tensors.append(inputs.input_values.squeeze(0).to(torch.float32))

    result = {
        "stem_paths": stem_paths,
        "stem_tensors": stem_tensors,
        "noise_tensors": noise_tensors
    }
    torch.save(result, cache_file)
    return result
```

```python

def run_training(model, train_ds, val_ds, device):
    if USE_WANDB:
        print("Starting WandB run...")
        wandb.init(
            project=WANDB_PROJECT,
            name=f"ast-train-{pd.Timestamp.now().strftime('%m%d-%H%M')}",
            config={
                "learning_rate": LEARNING_RATE,
                "epochs": EPOCHS,
                "batch_size": BATCH_SIZE,
                "model_id": MODEL_ID,
                "num_gpus": torch.cuda.device_count(),
                "num_variants": 1
            },
            reinit=True
        )

    training_args = TrainingArguments(
        output_dir=OUTPUT_DIR,
        save_strategy="epoch",
        eval_strategy="epoch",
        save_total_limit=2,
        learning_rate=LEARNING_RATE,
        per_device_train_batch_size=BATCH_SIZE,
        per_device_eval_batch_size=BATCH_SIZE,
        gradient_accumulation_steps=1, 
        num_train_epochs=EPOCHS,
        weight_decay=0.01,
        warmup_steps=200,
        lr_scheduler_type="cosine",
        label_smoothing_factor=0.1,
        load_best_model_at_end=True,
        metric_for_best_model="f1",
        greater_is_better=True,
        remove_unused_columns=False,
        gradient_checkpointing=False, 
        report_to="wandb" if USE_WANDB else "none",
        fp16=torch.cuda.is_available(),
        dataloader_num_workers=NUM_WORKERS,
        dataloader_pin_memory=True,

        logging_strategy="steps",
        logging_steps=10,
        do_eval=True,
        log_level="info",
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        compute_metrics=compute_metrics
    )

    trainer.train()
    trainer.save_model(os.path.join(OUTPUT_DIR, "final_best_model"))
    return trainer
```

```python
def run_submission_inference(model, feature_extractor, id2label, genres, device):
    test_csv = os.path.join(DATA_DIR, "test.csv")
    if not os.path.exists(test_csv):
        print("Test CSV not found. Skipping inference.")
        return

    test_df = pd.read_csv(test_csv)
    if not {'id', 'filename'}.issubset(test_df.columns):
        print("Missing required columns in test.csv")
        return

    results = []
    model.eval()

    with torch.no_grad():
        for _, row in tqdm(test_df.iterrows(), total=len(test_df), desc="Inference"):
            audio_path = os.path.join(DATA_DIR, row['filename'])
            try:
                if not os.path.exists(audio_path):
                    raise FileNotFoundError(audio_path)

                try:
                    waveform, sr = torchaudio.load(audio_path)
                except Exception as e:
                    print(f"Skipping corrupt file {audio_path}: {e}")
                    results.append({"id": row['id'], "genre": genres[0]})
                    continue

                if sr != feature_extractor.sampling_rate:
                    resampler = torchaudio.transforms.Resample(sr, feature_extractor.sampling_rate)
                    waveform = resampler(waveform)
                if waveform.shape[0] > 1:
                    waveform = waveform.mean(dim=0, keepdim=True)
                
                full_audio = waveform.squeeze(0).numpy()
                target_samples = int(feature_extractor.sampling_rate * TARGET_DURATION)
                
                audio_len = len(full_audio)
                if audio_len > target_samples:
                    # 3 windows: Start, Middle, End
                    offsets = [0, (audio_len - target_samples)//2, audio_len - target_samples]
                else:
                    offsets = [0]
                
                all_logits = []
                for offset in offsets:
                    window = full_audio[offset:offset+target_samples]
                    if len(window) < target_samples:
                        window = np.pad(window, (0, target_samples - len(window)))
                    
                    inputs = feature_extractor(window.astype(np.float32), sampling_rate=feature_extractor.sampling_rate, return_tensors="pt")
                    input_values = inputs.input_values.to(device)
                    outputs = model(input_values)
                    all_logits.append(outputs.logits)
                
                avg_logits = torch.stack(all_logits).mean(dim=0)
                pred = torch.argmax(avg_logits, dim=-1).item()
                results.append({"id": row['id'], "genre": id2label[pred]})

            except Exception as e:
                print(f"Error on {row.get('filename', 'unknown')}: {e}")
                results.append({"id": row['id'], "genre": genres[0]})

    submission_df = pd.DataFrame(results)
    submission_df = submission_df.sort_values(by='id').reset_index(drop=True)
    submission_df.to_csv(SUBMISSION_PATH, index=False)
    print(f"Submission saved: {SUBMISSION_PATH}")
```

```python
def main():
    seed = 42
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    
    hf_token = init_tokens()
    device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
    num_gpus = torch.cuda.device_count()
    print(f"Device: {device} | Number of GPUs: {num_gpus}")
    if num_gpus > 1:
        print(f"Using Multi-GPU Mode: Total Effective Batch Size = {BATCH_SIZE * num_gpus}")

    genre_data, noise_files = get_files(DATA_DIR)
    genres = sorted(list(genre_data.keys()))
    id2label = {i: g for i, g in enumerate(genres)}
    label2id = {g: i for i, g in enumerate(genres)}

    os.makedirs(OUTPUT_DIR, exist_ok=True)
    with open(os.path.join(OUTPUT_DIR, "label_map.json"), "w") as f:
        json.dump({"id2label": id2label, "label2id": label2id}, f)

    feature_extractor, model = setup_model(genres, id2label, label2id, device, hf_token)

    cache = prepare_simulation_data(genre_data, noise_files, feature_extractor)
    
    path_to_idx = {path: i for i, path in enumerate(cache["stem_paths"])}
    
    train_genre_indices = {}
    val_genre_indices = {}
    
    for g_idx, g_name in enumerate(sorted(genre_data.keys())):
        indices = [path_to_idx[p] for p in genre_data[g_name] if p in path_to_idx]
        
        if len(indices) < 2:
            train_genre_indices[g_idx], val_genre_indices[g_idx] = indices, []
        else:
            t, v = train_test_split(indices, test_size=0.1, random_state=42)
            train_genre_indices[g_idx], val_genre_indices[g_idx] = t, v

    train_ds = MashupSimulatorDataset(
        cache["stem_paths"], cache["stem_tensors"], train_genre_indices, cache["noise_tensors"], num_variants=1
    )
    val_ds = MashupSimulatorDataset(
        cache["stem_paths"], cache["stem_tensors"], val_genre_indices, cache["noise_tensors"], num_variants=2
    )

    print(f"Dataset ready. Training with {len(train_ds)} samples at full CPU/GPU speed.")
    trainer = run_training(model, train_ds, val_ds, device)

    print("Starting inference...")
    run_submission_inference(trainer.model, feature_extractor, id2label, genres, device)

    if USE_WANDB:
        wandb.finish()

```

```python
if __name__ == "__main__":
    main()
```
